```python
feature_category = ['gender', 'seniorcitizen', 'partner', 'dependents', 'phoneservice',
       'multiplelines', 'internetservice', 'onlinesecurity', 'onlinebackup',
       'deviceprotection', 'techsupport', 'streamingtv', 'streamingmovies',
       'contract', 'paperlessbilling', 'paymentmethod']

for feature in feature_category:
    plt.figure(figsize=(2,2))
    df.groupby([feature, 'churn']).size().unstack().plot(kind='bar', stacked = True)
    plt.xlabel(feature)
    plt.ylabel('count')
    plt.title(f'count of {feature} by churn')
    plt.xticks(rotation=45)
    plt.legend(title='Churn', loc='upper right')
    plt.show()
    
```

```python


def kdeplot(feature):
    plt.figure(figsize=(9, 4))
    plt.title('Kernel Density Estimation Plot for {}'.format(feature))
    ax0 = sns.kdeplot(df[df['churn']==0][feature], color='green',label='churn:No')
    ax1 = sns.kdeplot(df[df['churn']==1][feature], color='red',label='churn:Yes')
kdeplot('tenure')
kdeplot('monthlycharges')
kdeplot('totalcharges')
```
